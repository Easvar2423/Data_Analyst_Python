```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('workout_new.csv')
df
```

```python
plt.plot(df['Date'],df['Pulse'])
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (15, 5))
plt.plot(df['Date'], df['Pulse'], label = 'Pulse')
plt.plot(df['Date'], df['Duration'], label = 'Duration')
plt.xlabel('Date')
plt.ylabel('Duration & Pulse')
plt.title('Pulse and Duration Datewise')
plt.xticks(rotation=90)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (15, 5))
plt.plot(df['Date'], df['Pulse'], label = 'Pulse', color='g', marker ='o')
plt.plot(df['Date'], df['Duration'], label = 'Duration')
plt.xlabel('Date')
plt.ylabel('Duration & Pulse')
plt.title('Pulse and Duration Datewise')
plt.xticks(rotation=90)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (15, 5))
plt.plot(df['Date'], df['Pulse'], label = 'Pulse', color='g', marker ='o')
plt.plot(df['Date'], df['Duration'], label = 'Duration',color='r', marker ='o')
plt.xlabel('Date')
plt.ylabel('Duration & Pulse')
plt.title('Pulse and Duration Datewise')
plt.xticks(rotation=90)
plt.legend()
plt.grid()
plt.show()
```
