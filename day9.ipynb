```python
df = pd.read_csv('property-price.csv')
df
```

```python
X = df[['area','bedroom','age']]
Y = df[['price']]
```

```python
X
```

```python
Y
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[3800,2.5,10]])
```

```python
predictions = model.predict(X)
predictions
```

```python
df['Pred']= np.round(predictions,1)
df
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(df['price'],df['Pred'])*100
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('insurance.csv')
data
```

```python
X = data[['age']]
Y = data[['insurance']]
```

```python
X
```

```python
Y
```

```python
model = LogisticRegression()
model.fit(X,Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
predictions = model.predict(X)
data['Pred'] = predictions
data
```

```python
data[data['insurance'] != data['Pred']]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(data['insurance'], data['Pred'])
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('marks.csv')
df
```

```python
X = df[['Maths','CS']]
Y=df[['Result']]
```

```python
X
```

```python
Y
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X,Y)
```

```python
predictions = model.predict(X)
df['pred'] = predictions
df
```

```python
model.predict([[4,5]])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(df['Result'],df['pred'])
```
